````python
from os import listdir

import numpy as np
from cv2 import *
from sklearn.metrics import confusion_matrix
from sklearn.neighbors import KNeighborsClassifier

fldrs_name = ['Inside_City', 'Store', 'Mountain', 'Tall_Building', 'Open_Country', 'Highway', 'Kitchen', 'Coast',
              'Office', 'Bedroom', 'Suburb', 'Livingroom', 'Street', 'Industrial', 'Forest']

for size in range(3, 20):
    train_X = []
    train_Y = []


    def training(path, index):
        for file in listdir(path):
            if file.endswith(".jpg"):
                im = imread(path + file)
                im = resize(im, (size, size))
                x = np.array(im).reshape((-1))
                train_X.append(x)
                train_Y.append(index)
            else:
                continue


    for i in range(len(fldrs_name)):
        filename = fldrs_name[i]
        training('Data/Train/' + filename + '/', i)

    test_X = []
    true_Y = []


    def test(path, index):
        for file in listdir(path):
            if file.endswith(".jpg"):
                im = imread(path + file)
                im = resize(im, (size, size))
                x = np.array(im).reshape((-1))
                test_X.append(x)
                true_Y.append(index)
            else:
                continue


    for i in range(len(fldrs_name)):
        filename = fldrs_name[i]
        test('Data/Test/' + filename + '/', i)

    for k in range(4):
        for l in range(2):
            metr = 'l' + str(l + 1)
            neigh = KNeighborsClassifier(n_neighbors=2 * k + 1, metric=metr)
            neigh.fit(train_X, train_Y)
            test_Y = neigh.predict(test_X)

            C = confusion_matrix(true_Y, test_Y)

            correct = 0
            for i in range(15):
                correct += C[i, i]
            print('size: ' + str(size), ' kNN: ' + str(2 * k + 1), ' metr: ' + metr)
            print(correct / 15)
````

در این قسمت از سوال ابتدا روی سایزی که عکس را به آن ریسایز می‌کنیم فور می‌زنیم و برای مقادیر 3×3 تا 19×19 حساب می‌کنیم.
در تابع ترینینگ آدرس فایل را می‌گیریم و عکس را می‌خوانیم و آن را به اندازه‌ی 
$size×size$
ریسایز می‌کنیم و به صورت یک بردار 
$size^2$تایی
ریشیپ می‌کنیم و به ماتریس 
$train\_X$
اضافه می‌کنیم. هم‌چنین اندیس را هم در ورودی تابع می‌گیریم و به ماتریس 
$train\_Y$
آن را می‌افزاییم. 
این کار را برای تمامی نوع داده‌ها انجام می‌دهیم و یک ماتریس از فیچر وکتور‌های عکس‌های آموزش که بردار ریسایز شده‌ی آن‌هاست می‌سازیم.
سپس برای تمامی عکس‌های تست نیز فیچر وکتور هر عکس را به صورت قبلی حساب می‌کنیم و به ماتریس 
$test\_X$
و مقدار اندیس که نشان‌دهنده نوع تصویر است را به ماتریس 
$test\_Y$
اضافه می‌کنیم.

در قسمت آخر برای سنجیدن حدس و نتیجه‌ای که توسط روش 
$kNN$
روی مقادیر مختلف $k$
و دو نوع نرم $l1,l2$
فور می‌زنیم و با 


```
 neigh = KNeighborsClassifier(n_neighbors=2 * k + 1, metric=metr)
```

کلاس‌بندی می‌کنیم داده‌ها را و سپس با استفاده از همین داده‌های آموزش را به هم فیت می‌کنیم.
سپس 
$test\_Y$
را با استفاده از الگوریتم 
$kNN$
حدس می‌زنیم.
