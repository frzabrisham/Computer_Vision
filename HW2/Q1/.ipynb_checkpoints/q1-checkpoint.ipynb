```python
import numpy as np
from cv2 import *
from numpy.linalg import inv

path = 'Frames4'
path450 = path + '/f0450.jpg'
path270 = path + '/f0270.jpg'
path630 = path + '/f0630.jpg'
path90 = path + '/f0090.jpg'
path810 = path + '/f0810.jpg'
path1 = path + '/f0001.jpg'

img450, im450 = imread(path450), imread(path450)
img270 = imread(path270)
img630 = imread(path630)
img90 = imread(path90)
img810 = imread(path810)
img1 = imread(path1)

h = int(img450.shape[0])
w = int(img450.shape[1])
# print(h, w)
sift = cv2.xfeatures2d.SIFT_create()

FLANN_INDEX_KDTREE = 0
index_params = dict(algorithm=FLANN_INDEX_KDTREE, trees=5)
search_params = dict(checks=50)
flann = FlannBasedMatcher(index_params, search_params)

kp450, des450 = sift.detectAndCompute(img450, None)
kp270, des270 = sift.detectAndCompute(img270, None)
kp630, des630 = sift.detectAndCompute(img630, None)
kp90, des90 = sift.detectAndCompute(img90, None)
kp810, des810 = sift.detectAndCompute(img810, None)
kp1, des1 = sift.detectAndCompute(img1, None)

image_folder = os.fsencode(path)

filenames = []
for file in os.listdir(image_folder):
    filename = os.fsdecode(file)
    if filename.endswith('.jpg'):
        filenames.append(filename)

filenames.sort()
```

```python
def matching(matches, kp_main, kp_min):
    matched_main, matched = [], []
    thresh = 0.6
    for match in matches:
        p1 = kp_main[match[0].queryIdx].pt
        p2 = kp_min[match[0].trainIdx].pt

        if match[0].distance < thresh * match[1].distance:
            matched_main.append([p1])
            matched.append([p2])

    return matched_main, matched


def corner(H):
    t0 = H @ [0, 0, 1]
    t0 /= t0[2]
    t1 = H @ [w - 1, 0, 1]
    t1 /= t1[2]
    t2 = H @ [0, h - 1, 1]
    t2 /= t2[2]
    t3 = H @ [w - 1, h - 1, 1]
    t3 /= t3[2]
    return t0, t1, t2, t3


def dim(H):
    t0, t1, t2, t3 = corner(H)
    x = max(max(t0[0], t1[0], t2[0]), t3[0]) - min(min(t0[0], t1[0], t2[0]), t3[0])
    y = max(max(t0[1], t1[1], t2[1]), t3[1]) - min(min(t0[1], t1[1], t2[1]), t3[1])
    # p = [int(x[0][0]), int(y[0][0])]
    return [int(x[0][0]), int(y[0][0])]


def trans(H):
    t0, t1, t2, t3 = corner(H)
    x = min(min(t0[0], t1[0], t2[0]), t3[0])
    y = min(min(t0[1], t1[1], t2[1]), t3[1])
    p = [x[0][0], y[0][0]]
    t = ([[1, 0, -p[0] + 50], [0, 1, -p[1] + 50], [0, 0, 1]])
    return t


def masking(im):
    mask = np.zeros((im.shape[0], im.shape[1]))
    for i in range(im.shape[0]):
        for j in range(im.shape[1]):
            if im[i][j][0] != 0 and im[i][j][1] != 0 and im[i][j][2] != 0:
                # print(img[i][j])
                mask[i][j] = 1
    return mask


# im1 is under im2
def merge_im(im1, im2):
    im = np.zeros((im1.shape[0], im1.shape[1], 3))
    mask1 = masking(im1)
    mask2 = masking(im2)
    for i in range(im1.shape[0]):
        for j in range(im1.shape[1]):
            if mask1[i][j] == 1 and mask2[i][j] == 0:
                im[i][j] = im1[i][j]
            elif mask2[i][j] == 1:
                im[i][j] = im2[i][j]
    return im


def writeim(arr, num, name, folder=''):
    imwrite(folder + 'res' + num + '-' + name + '.jpg', arr)


def find_hom(des_ref, des_minor, kp_ref, kp_minor):
    matches = flann.knnMatch(des_ref, des_minor, k=2)
    matched_1, matched_2 = matching(matches, kp_ref, kp_minor)
    H, status = findHomography(np.array(matched_2), np.array(matched_1), RANSAC, maxIters=10000)

    return H
```

```python
# find homography for key frames
I_450 = find_hom(des_ref=des450, des_minor=des450, kp_ref=kp450, kp_minor=kp450)

h_270 = find_hom(des_ref=des450, des_minor=des270, kp_ref=kp450, kp_minor=kp270)

h_630 = find_hom(des_ref=des450, des_minor=des630, kp_ref=kp450, kp_minor=kp630)

h_90 = find_hom(des_ref=des270, des_minor=des90, kp_ref=kp270, kp_minor=kp90)
h_90 = h_90 @ h_270

h_810 = find_hom(des_ref=des630, des_minor=des810, kp_ref=kp630, kp_minor=kp810)
h_810 = h_810 @ h_630

h_1 = find_hom(des_ref=des90, des_minor=des1, kp_ref=kp90, kp_minor=kp1)
h_1 = h_1 @ h_90
# %%
# warped rectangle

a, b = 1 / 3, 2 / 3
pts = np.array([a * w, a * h, a * w, b * h, b * w, b * h, b * w, a * h], np.int32)
pts = pts.reshape((-1, 1, 2))

color = (150, 200, 50)
thickness = 4

rect_450 = polylines(im450, [pts], isClosed=True, color=color, thickness=thickness)
writeim(arr=rect_450, num='01', name='450-rect')

H270_inv = inv(h_270)
rect = polylines(np.zeros((h, w, 3)), [pts], isClosed=True, color=color, thickness=thickness)
warped_rect = warpPerspective(rect, H270_inv, (w, h))

rect_270 = merge_im(img270, warped_rect)
writeim(arr=rect_270, num='02', name='270-rect')

# %%
# panorama with 450 and 270

T_270 = trans(h_270)
H270 = T_270 @ h_270
dim270 = (dim(H270)[0] + 200, dim(H270)[1] + 200)
warped_270 = warpPerspective(img270, H270, dim270)

I_450 = T_270 @ I_450
tran_450 = warpPerspective(img450, I_450, dim270)

panorama_270_450 = merge_im(warped_270, tran_450)
writeim(arr=panorama_270_450, num='03', name='270-450-panorama')
```
